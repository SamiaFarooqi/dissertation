```python
#importing the primary libraries
import numpy as np
import pandas as pd
```

```python
#importing libraries for visualizing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tc_data=pd.read_csv('crash_reporting_data.csv')
tc_data.head()
```

```python
tc_data.info()
```

```python
tc_data["Crash Date/Time"] = pd.to_datetime(tc_data["Crash Date/Time"])
```

```python
tc_data.head(5)
```

```python
tc_data["CRASH YEAR"] = tc_data["Crash Date/Time"].dt.year
tc_data["CRASH MONTH"] = tc_data["Crash Date/Time"].dt.month
tc_data["CRASH HOUR"] = tc_data["Crash Date/Time"].dt.hour
tc_data["CRASH WEEK"] = tc_data["Crash Date/Time"].dt.strftime('%a')
```

```python
tc_data.head(5)
```

Handling the null Values

```python
tc_data.columns
```

```python
#Finding the columns with textx values
text_columns = []
for column in tc_data.columns:
    if tc_data[column].dtype == 'object':
        text_columns.append(column)

print("Text Columns:", text_columns)
```

```python
#Dropping the columns with useless values
columns_to_drop = ['Agency Name', 'Road Name', 'Cross-Street Name', 'Driver Distracted By', 'Drivers License State', 'Vehicle Make', 'Vehicle Model']
tc_data.drop(columns=columns_to_drop, inplace=True)
```

```python
tc_data.shape
```

```python
# Checking for null values in each column
null_columns = tc_data.columns[tc_data.isnull().any()]

# Getting the number of null values in each column
null_counts = tc_data[null_columns].isnull().sum()

# Printing columns with their respective null counts
for column, null_count in zip(null_counts.index, null_counts):
    print(f"Column '{column}' has {null_count} null values.")
```

```python
# Drop rows with null values
tc_data.dropna(axis=0, inplace=True)
```

```python
# Map categories to binary values
binary_encoded = tc_data['Injury Severity'].map({'NO APPARENT INJURY': 1,
                                                 'SUSPECTED MINOR INJURY': 1,
                                                 'SUSPECTED SERIOUS INJURY': 1,
                                                 'POSSIBLE INJURY': 1,
                                                 'FATAL INJURY': 0})


binary_encoded = binary_encoded.astype(int)
tc_data['Injury Severity Encoded'] = binary_encoded

```

```python
tc_data['Crash Date/Time'] = tc_data['Crash Date/Time'].map(pd.Timestamp.timestamp).astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize an empty list to store categorical column names
cate = []

# Identify categorical columns
for col in tc_data.columns:
    if tc_data[col].dtype == 'object':
        cate.append(col)

# Initialize LabelEncoder
label_encoders = {}

# Apply label encoding to each categorical column
for col in cate:
    label_encoders[col] = LabelEncoder()
    tc_data[col] = label_encoders[col].fit_transform(tc_data[col])

# Display the transformed dataset
tc_data.head()
```

Modelling


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X = tc_data.drop(columns=['Injury Severity'])
y = tc_data['Injury Severity']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Scale features for neural network models
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from xgboost import XGBClassifier

# Initialize the XGBoost classifier
xgb_classifier = XGBClassifier()

# # Training
xgb_classifier.fit(X_train_scaled, y_train)
y_pred = xgb_classifier.predict(X_test_scaled)

# Evaluate the classifier
from sklearn.metrics import accuracy_score, classification_report

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.neural_network import MLPClassifier

# Initialize the MLP classifier
mlp_classifier = MLPClassifier()

# Training
mlp_classifier.fit(X_train_scaled, y_train)
y_pred_mlp = mlp_classifier.predict(X_test_scaled)

# Evaluate the classifier
from sklearn.metrics import accuracy_score, classification_report

# Calculate accuracy
accuracy_mlp = accuracy_score(y_test, y_pred_mlp)
print("Accuracy (MLP):", accuracy_mlp)

# Generate classification report
print("Classification Report (MLP):")
print(classification_report(y_test, y_pred_mlp))
```

```python
# Define the models and their accuracies
models = ['XGBoost', 'Neural Network']
accuracies = [accuracy, accuracy_mlp]

# Plotting
plt.figure(figsize=(8, 6))
sns.barplot(x=models, y=accuracies, palette='Set2')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.ylim(0, 1)  # Limit y-axis to range between 0 and 1 for accuracy
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
# Get classification reports for XGB and MLP classifiers
xgb_report = classification_report(y_test, y_pred, output_dict=True)
mlp_report = classification_report(y_test, y_pred_mlp, output_dict=True)

# Extract metrics from classification reports
metrics = ['precision', 'recall', 'f1-score']
xgb_metrics = {metric: xgb_report['weighted avg'][metric] for metric in metrics}
mlp_metrics = {metric: mlp_report['weighted avg'][metric] for metric in metrics}

# Plot comparison
plt.figure(figsize=(10, 6))
sns.barplot(x=list(xgb_metrics.keys()), y=list(xgb_metrics.values()), color='blue', alpha=0.5, label='XGB')
sns.barplot(x=list(mlp_metrics.keys()), y=list(mlp_metrics.values()), color='red', alpha=0.5, label='MLP')
plt.xlabel('Metrics')
plt.ylabel('Score')
plt.title('Comparison of XGB and MLP Classifiers')
plt.legend()
plt.show()

# Plot confusion matrices
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap='Blues', fmt='d', cbar=False)
plt.title('Confusion Matrix - XGB Classifier')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

plt.subplot(1, 2, 2)
sns.heatmap(confusion_matrix(y_test, y_pred_mlp), annot=True, cmap='Reds', fmt='d', cbar=False)
plt.title('Confusion Matrix - MLP Classifier')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

plt.tight_layout()
plt.show()
```
